```python
import json
import os
import re
```

```python
val_imgNamelist = []
for root, _, files in os.walk("../data/ccksdata/val"):
    val_imgNamelist = files
```

```python
val_imgNamelist
```

```python
pjson = json.load(open("../data/ccksdata/problems.json"))
pjson
```

```python
pjson["727"]["diagram_path"]
```

```python
pattern = re.compile(r"\\[1-9][0-9]*.png$")
result = re.search(pattern, pjson["727"]["diagram_path"])
result.group()[1:]
```

```python
val_pid_list = []
for n in val_imgNamelist:
    for i, data in pjson.items():
        pattern = re.compile(r"\\[1-9][0-9]*.png$")
        result = re.search(pattern, data["diagram_path"])
        if result.group()[1:] == n:
            val_pid_list.append(i)
```

```python
len(val_pid_list)
```

```python
train_imgNamelist = []
for root, _, files in os.walk("../data/ccksdata/train"):
    train_imgNamelist = files
```

```python
train_imgNamelist
```

```python
train_pid_list = []
for n in train_imgNamelist:
    for i, data in pjson.items():
        pattern = re.compile(r"\\[1-9][0-9]*.png$")
        result = re.search(pattern, data["diagram_path"])
        if result.group()[1:] == n:
            train_pid_list.append(i)
```

```python
len(train_pid_list)
```

```python
train_pid_list
```

```python
test_pid_list = []
for i in range(664):
    test_pid_list.append(str(i + 1997))
test_pid_list
```

```python
pid_split = dict()
pid_split["train"] = train_pid_list
pid_split["val"] = val_pid_list
pid_split["test"] = test_pid_list
```

```python
json.dump(pid_split, open("../data/ccksdata/pid_splits.json", "w"))
```
